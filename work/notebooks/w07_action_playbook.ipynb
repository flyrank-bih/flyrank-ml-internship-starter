## 1. Ranked actions + reason codes

*The queue: what to do first, and why, in words a human trusts.*

This queue prioritizes content updates based on the model's predicted decay risk and historical traffic volume. The recommendations are directional and meant to support decision-making, not automate it.

```python
import pandas as pd
import os

# Placeholder: Replace this with your actual scored dataset from Week 8/9
# Ensure your actual data contains no client names, URLs, or private queries.
mock_data = {
    'page_id': ['1042', '8831', '9012', '3321', '4451'],
    'predicted_risk_score': [0.89, 0.82, 0.75, 0.60, 0.12],
    'historical_traffic': [15000, 2200, 18000, 500, 45000],
    'primary_action': ['Rewrite', 'Consolidate', 'Rewrite', 'Archive', 'Monitor'],
    'reason_code': [
        'High traffic, high decay risk',
        'Low traffic, topical overlap',
        'High traffic, moderate decay risk',
        'Low traffic, high decay risk',
        'High traffic, stable performance'
    ]
}

action_queue = pd.DataFrame(mock_data)

# Sort by risk score (descending) to prioritize highest risk
ranked_queue = action_queue.sort_values(by='predicted_risk_score', ascending=False).reset_index(drop=False)

display(ranked_queue)
```

## 2. Intended use and limits

*Who uses this, for what — and where it stops being valid.*


*Who uses this, for what — and where it stops being valid.*

* **Intended Users:** Content managers, SEO strategists, and editorial leads.
* **Intended Use:** A decision-support tool to prioritize which existing pages require auditing, rewriting, or archiving based on risk scores.
* **Limits:**
  1. The model assumes historical traffic and decay patterns will hold; it cannot predict sudden external market shifts.
  2. The risk score is a measured probability based on past features, not a guaranteed indicator of future page failure.

## 3. Human review + the no-go list

*What a person must check before acting. What should never be automated.*


*What a person must check before acting. What should never be automated.*

**Human review rules:**
* A domain expert must verify the content's actual relevance and accuracy before executing a "Rewrite" action.
* Traffic drops must be cross-referenced with seasonal trends by a human reviewer before deciding to archive.

**The No-Go List (What should NOT be automated):**
* Automated deletion or unpublishing of any page without human sign-off.
* Automated generation and publishing of new content to replace flagged pages.
* Using these scores to evaluate the performance of individual human authors.

## 4. Monitoring / retrain triggers

*What would tell you the recommendations went stale?*


*What would tell you the recommendations went stale?*

**Light monitoring and retrain triggers:**
* **Feature Drift:** If the distribution of incoming data (e.g., baseline traffic) shifts by more than 20% from the training baseline.
* **Performance Degradation:** If the precision of our top-decile predictions drops below our established baseline over a rolling 30-day window.
* **Time Trigger:** If 90 days have passed since the last model training run.
