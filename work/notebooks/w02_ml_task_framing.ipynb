## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

**Task Type:** Ranking / Priority Scoring (via Binary Classification probabilities)

**Why?**
The goal for FlyRank's content team is to identify which existing pages are decaying and need an immediate refresh. Pure classification (declining vs. not declining) isn't enough because human editors have limited bandwidth (e.g., they can only rewrite 50 pages a week). We frame this as a **ranking task** where we use the probabilities from a classification model to score and order the pages, ensuring the most urgent opportunities appear at the top of the queue.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

**Target / Proxy:** `is_declining_label` (Proxy)

**Where it comes from:**
It comes from an **observed outcome** in the historical search data. Specifically, it is derived from the `trend_direction` feature (`trend_direction == 'down'`). We use this as a proxy to indicate that a page is losing its search relevance or demand over the evaluated time window.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

**Success Metric:** **Precision@50**

**What number means 'good'?**
Since the editorial team will likely only review the top results, overall accuracy doesn't matter as much as the top of the list. Our fixed hand-written rule (baseline) achieves a Precision@50 of ~0.24. A 'good' number for the ML model would be anything significantly higher, such as **> 0.70**. This would mean that out of the top 50 pages recommended for a refresh, at least 35 are genuinely declining, saving the editors massive amounts of wasted time.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

**Why ML beats a fixed rule:**
Content decay is driven by a tangled mix of signals. A fixed business rule (like `IF content_age > 180 AND impressions < 500`) relies on arbitrary human thresholds and misses nuances. ML beats the fixed rule because it can dynamically weigh the non-linear interactions between dozens of variables—such as how CTR, average position, word count, and engagement rate all interact together—to find hidden patterns of decay that simple `if-statements` cannot capture.
