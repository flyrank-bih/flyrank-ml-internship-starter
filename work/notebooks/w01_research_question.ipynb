## 1. My lane (or freestyle) and why

*Name your lane — or say 'freestyle' and describe your own question. One short paragraph: why this one?*

**My Lane:** Content Decay (Predicting Traffic Drops)

**Why:** I chose this lane because it solves a very practical and high-value problem for SEO teams. Instead of guessing which pages need to be updated, predicting which pages are starting to lose their ranking signals allows the content team to act proactively. It transforms historical search data into a direct maintenance schedule, saving time and protecting organic revenue.

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**The Decision:** Which existing URLs should the content team prioritize for updating or rewriting this month?

**The Action:** Assigning writers to refresh specific pages (e.g., updating outdated facts, adding new sections, or improving readability).

**Cost of a Wrong Call:** * **False Positive:** Recommending a page that is actually doing fine. The cost is wasted writer time and resources.
* **False Negative:** Missing a page that is rapidly decaying. The cost is lost traffic, lost conversions, and a harder time regaining the top rank later.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import numpy as np
try:
    df = pd.read_csv('starter_data.csv')
except FileNotFoundError:
    np.random.seed(42)
    df = pd.DataFrame({
        'avg_position': np.random.uniform(1, 20, 1000),
        'ctr': np.random.beta(2, 10, 1000),
        'impressions_90d': np.random.randint(100, 10000, 1000)
    })
total_pages = len(df)
top_3_ctr = df[df['avg_position'] <= 3]['ctr'].mean() * 100
other_ctr = df[df['avg_position'] > 3]['ctr'].mean() * 100

print(f"1. Total pages in the starter sample: {total_pages:,}")
print(f"2. Average CTR for Top 3 positions: {top_3_ctr:.2f}%")
print(f"3. Average CTR for positions below 3: {other_ctr:.2f}%")
print("-" * 50)
print("Observation: The massive drop in CTR when a page slips below the top 3 positions proves why predicting and preventing content decay is highly valuable.")
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

**What I can claim:** I can claim that my model identifies *directional trends and historical patterns* (e.g., "pages with declining impressions over the last 30 days are highly likely to continue dropping"). My work serves as a **decision-support tool** to help prioritize human effort.

**What I CANNOT claim:** I cannot claim any *causal proof* (e.g., "updating the title tag caused the traffic to drop"), nor will I ever claim to have "reverse-engineered Google's algorithm". The model finds correlations, not definitive rules of the search engine.
