## 3. Quick look at the data (2-3 real numbers)

```python
# Load starter dataset
import pandas as pd
import numpy as np

df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')

# Show basic info
print(f"Total pages in dataset: {len(df):,}")
print(f"\nColumns available: {df.columns.tolist()}")
print(f"\nMemory usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# Number 1: Visible pages with meaningful traffic
visible = df[df['impressions_90d'] > 500]
print(f"\n1. Pages with 500+ impressions in 90 days: {len(visible):,} ({len(visible)/len(df)*100:.1f}%)")

# Number 2: Position distribution
print(f"\n2. Average position distribution:")
print(f"   Min: {df['avg_position'].min():.1f}")
print(f"   Max: {df['avg_position'].max():.1f}")
print(f"   Mean: {df['avg_position'].mean():.1f}")
print(f"   Median: {df['avg_position'].median():.1f}")

# Number 3: Movement
print(f"\n3. Trend direction breakdown:")
print(df['trend_direction'].value_counts())
```

**Why this lane looks worth exploring:**

**1. Scale of opportunity:** {X} pages have 500+ impressions in 90 days — that's enough volume to find meaningful patterns. Even small improvements across these pages could make a big impact.

**2. Wide variation:** Average position ranges from {min} to {max}, with a median of {median}. This variation means there's room to understand what drives better positions.

**3. Real movement:** {X}% of pages are trending down — there's actual churn happening, and understanding why matters.
